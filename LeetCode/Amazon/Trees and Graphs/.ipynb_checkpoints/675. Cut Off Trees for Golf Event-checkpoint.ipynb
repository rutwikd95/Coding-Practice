```python
def cut_trees(forest):
    if forest[0][0] == 0:
        return -1
    steps = 0 
    stack = [(0,0)]
    visited = []
    while stack :
        cur_row, cur_col = stack.pop()
#         visited[(cur_row, cur_col)] = True
        visited.append((cur_row , cur_col))
        if forest[cur_row][cur_col] == 0:
            return
        if forest[cur_row][cur_col] >= 1:
            forest[cur_row][cur_col] = 1
            new_row , new_col = get_min(forest, cur_row, cur_col, visited)
            stack.append((new_row, new_col))
            steps += 1
    
    return steps
            
            
def get_min(forest, i , j, visited):
#     visited = []
    directions = [(0,1) , (1,0) , (-1,0) , (0,-1)]
    next_list = []
    
    for d in directions:
        row , col = i + d[0] , j + d[1]
        if is_valid(forest, row, col) and forest[row][col] != 0 and (row, col) not in visited:
            next_list.append((forest[row][col] , row , col))
            
    print(next_list)
    min_val = 99999999999
    min_row = 0
    min_col = 0
    for val in next_list:
        if val[0] <= min_val:
            min_val = val[0]
            min_row = val[1]
            min_col = val[2]
    return min_row , min_col
     
def is_valid(forest, i , j):
    m , n = len(forest) , len(forest[0])
    if i < 0 or i >= m or  j < 0 or j >= n:
        return False
    return True
            
            
    
    
        
```
