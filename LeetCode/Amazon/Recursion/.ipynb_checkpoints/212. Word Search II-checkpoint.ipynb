```python
def wordsearch_2(board, words):
    ans = []
    for word in words:
#         first = word[0]
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == word[0]:
                    current = ""
                    word_index = 0
                    ans = dfs(current , ans, board, i , j, word, word_index)
                    
    return ans

def dfs(current, ans, board , i , j, word, word_index):
    print("Current : ", current)
#     print(ans)
    if len(current) == len(word):
        print("Appended")
        ans.append(current)
    else:
        print("Word at Word Index : " ,word[word_index] )
        print("Board : ",board[i][j] )
        if board[i][j] == word[word_index] :
            if is_valid(board , i+1 , j , word_index,word):
                print("enter 1")
                dfs(current + word[word_index] , ans , board, i+1 , j ,word , word_index + 1)
            if is_valid(board , i-1 , j, word_index,word):
                print("enter 2")
                dfs(current + word[word_index] , ans , board, i-1 , j ,word, word_index + 1)
            if is_valid(board , i , j+1, word_index,word):
                print("enter 3")
                dfs(current + word[word_index] , ans , board, i , j+1 ,word, word_index + 1)
            if is_valid(board , i , j-1, word_index,word):
                print("enter 4")
                dfs(current + word[word_index] , ans , board, i , j-1 ,word, word_index + 1)
        
    return ans
    
def is_valid(board, i , j, word_index,word):
    if i < 0 or  i > len(board) or j < 0 or j > len(board[0]) or word_index > len(word):
        return False
    return True
                
                
```

```python
board = [['o','a','a','n'],
         ['e','t','a','e'],
         ['i','h','k','r'],
         ['i','f','l','v']]


words = ["oath","pea","eat","rain"]

wordsearch_2(board, words)
```
