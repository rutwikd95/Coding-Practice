```python
def small_mul(n):
    flag = False
#     allowed_chars = set("01")
    stack = 1
#     i = 1
    while flag == False:
        if stack % n == 0 :
            ans = stack
            return ans
            flag = True
            break;
        else:
            stack = increm_stack(stack)
            print("stack : ",stack)
            
            
def increm_stack(stack):
    stack = str(stack)
    if stack[len(stack) - 1] == 1:
        stack = stack + str(0)
        stack = int(stack)
        print("increm stack 0 : ", stack)

    else:
        stack = stack[ : len(stack) - 1 ]
        stack = stack + "1"
        stack = int(stack)
        print("increm stack 1 : ", stack)
    return stack

```
